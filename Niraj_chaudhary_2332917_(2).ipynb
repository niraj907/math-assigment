````python
##  Question Number two
## Niraj Chaudhary
## 2332917

#creating the primary recursion function, which has two parameters.
def Solutiontwo(y,x):
  #whenever the lowest number, "x," and the greatest number, "y," are equal.
  if (y <= x):
    #It will merely give you back the lowest number, "X." This return will now prevent the recursion..
    return x
  #each time the first if statement is false.
  return y + Solutiontwo(y-1,x) ##This operates in a similar manner to the factorial. But rather than multiplying the number, we add it. The highest input serves as the first parameter for the y parameter. The same function is then applied to the addition. Recursion is the term for this.
##The main function will now continue to loop until the value of y coincides with the lowest value of the 'x' parameter.
     
#requesting input from the user for the lesser and larger number.
while True:
  try:
    num1 = int(input("Enter the first number: ")) 
    num2 = int(input("Enter the second number: ")) 
    break
  except ValueError:
    print("Please input numbers only.")

if (num1 < num2):
  #From User Input
  user_input= (Solutiontwo(num2,num1))
  print("From User Input: ",user_input)
else:
  print("First num must be smaller than the second. ")
#based on my university ID (i.e. = 2332917 = '2' 3329 '17')
student_number = Solutiontwo(17,2)
print("From Student Number: ",student_number)

#Based on my Height (i.e = 6ft )
MY_height = Solutiontwo(6,-12)
print("MY Height: ",MY_height)

````

````python
## Question Number Three
## Niraj Chaudhary
## 2332917

import matplotlib.pyplot as plt #importing library

plt.title("Niraj Chaudhary -  Solution of Three Number") # title
plt.xlabel("X - Axis") #X axis definition
plt.ylabel("Y - Axis") #Y axis definition

#For the first line, function:
def one(x):
    return 3*x*x

#For the second line,function:
def two(x):
    return (4*x)-3

#The main function for calculating points.
def calc():
  a = [] #Creating an empty list for a
  b = [] #Creating an empty list for b
  c = [] #Creating an empty list for c
  d = [] #Creating an empty list for d
  for i in range(-5,16): #using the for loop's -5 to 15 range as shown by the question.
     a.append(i) #adding the numbers -5 to 15 to the a list.
     b.append(one(i)) #adding the function one function's returned value to the b list.
     c.append(i) #adding the numbers -5 to 15 to the c list.
     d.append(two(i)) #adding the function one function's returned value to the d list.
  plt.plot(a,b,c,d,marker="o") #utilizing the 'o' marker to indicate the final points of both points on the same graph.
  plt.grid() #according to the requirements of the query, displaying the grid to the graph.
  plt.show() #presenting or demonstrating the final graph.
calc() #Calling the function
      
````
